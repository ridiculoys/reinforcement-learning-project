```python
import pandas as pd

def load_data(filename):
    fp = open(filename, "r")
    data = [[int(j.strip()) for j in i.split(" ")] for i in fp]
    df =  pd.DataFrame(data=data, columns=["iter", "p1_win", "p2_win", "draw", "p1_start", "p2_start"])
    fp.close()

    return df
```

```python
def plot_data(title, df):
    data = pd.DataFrame({
        'p1': df.loc[:, "p1_win"].tolist(),
        'p2': df.loc[:, "p2_win"].tolist(),
        'draw': df.loc[:, "draw"].tolist()
    }, index=df.loc[:, "iter"].tolist())
    
    print('p1', df.loc[:, "p1_win"].tolist())
    print('p2', df.loc[:, "p2_win"].tolist())
    print('drwas', df.loc[:, "draw"].tolist())
    print('iter', df.loc[:, "iter"].tolist())

    print(data)

    lines = data.plot.line(title=title)
```

```python
random_v_random = load_data("1_random_v_random.txt")
q_v_random = load_data("2_q_v_random.txt")
trained_v_random = load_data("3_trained_v_random.txt")
trained_v_q = load_data("4_1_trained_v_q.txt")
trained_v_trained = load_data("5_trained_v_trained.txt")
```

```python
data = [random_v_random, q_v_random, trained_v_random, trained_v_q, trained_v_trained]
for df in data:
    print(df)
```
